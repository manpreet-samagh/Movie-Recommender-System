```python
import pandas as pd
```

```python
r_cols = ['user_id','movie_id','rating']
ratings = pd.read_csv('u.data.csv', sep='\t',names=r_cols,usecols=range(3))
m_cols = ['movie_id','title']
ratings.head()
```

```python
movies = pd.read_csv('u.item.csv',sep='|',names=m_cols,usecols=range(2),encoding = "ISO-8859-1")
```

```python
ratings=pd.merge(movies,ratings)
ratings.head()
```

```python
movieRatings = ratings.pivot_table(index=['user_id'],columns=['title'],values='rating')
movieRatings.head()
```

```python
youngFrankensteinRatings= movieRatings['Young Frankenstein (1974)']
youngFrankensteinRatings.head()
```

```python
similarMovies = movieRatings.corrwith(youngFrankensteinRatings)
similarMovies = similarMovies.dropna()
df = pd.DataFrame(similarMovies)
df.head()
```
